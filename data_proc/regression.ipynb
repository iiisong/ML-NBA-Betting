```python
import numpy as np
from sklearn import linear_model, model_selection, metrics
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Example linear regression model
reg = linear_model.LinearRegression()
df_basic = pd.read_csv("games_basic_master.csv")

X = df_basic[['TEAM1_ID', 'TEAM2_ID', 'away_prob', 'away_spread', 'away_total']]
y = df_basic['TEAM1_SCORE']

# test_size is proportion of the dataset to include in the test split
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=42)
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)

coefficients = reg.coef_
intercept = reg.intercept_
print("Coefficients:", coefficients)
print("Intercept:", intercept)
mse = metrics.mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
r2 = metrics.r2_score(y_test, y_pred)
print(f'R-squared: {r2}')

```

```python
#Visualize
plt.scatter(y_test, y_pred)
plt.xlabel("ACTUAL TEAM1 SCORE")
plt.ylabel("PREDICTED TEAM1 SCORE")
plt.title("ACTUAL VS PREDICTED TEAM1 SCORE")
plt.show()

residuals = y_test - y_pred
plt.scatter(y_test, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel("ACTUAL TEAM1 SCORE")
plt.ylabel("PREDICTED TEAM1 SCORE")
plt.title("RESIDUAL PLOT")
plt.show()
```
